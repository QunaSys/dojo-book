```python
import sympy
from sympy import simplify, init_printing
from sympy.physics.quantum import represent, qapply
from sympy.physics.quantum.qubit import Qubit, measure_partial
from sympy.physics.quantum.gate import H, CGateS
print(sympy.__version__)
# уГЩуВпуГИуГлуВДшбМхИЧуВТуБНуВМуБДуБлшбичд║уБЩуВЛуБЯуВБуБощЦвцХ░уВТхоЯшбМ
init_printing()
```

```python
ctrlH = CGateS(0, H(1))
represent(ctrlH, nqubits=2)
```

```python
state = Qubit('00') # уВвуГАуГЮуГ╝уГлуГЖуВ╣уГИуБохИЭцЬЯчК╢цЕЛ
state_before_meas = qapply(H(0) * ctrlH * H(0) * state)
state_before_meas
```

```python
measured_state_and_probability_zero = measure_partial(state_before_meas,(0,))[0]
simplify(measured_state_and_probability_zero[1])
```

```python
measured_state_zero = measured_state_and_probability_zero[0]
simplify(measured_state_zero)
```

```python
simplify( qapply( H(1) * measured_state_zero ) )
```
