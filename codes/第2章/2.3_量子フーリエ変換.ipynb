```python
import sympy
from sympy import simplify, init_printing
from sympy.physics.quantum import represent, qapply
from sympy.physics.quantum.qubit import Qubit, measure_partial
from sympy.physics.quantum.gate import Z, S, T, H, CGateS, SWAP
# ベクトルや行列をきれいに表示するための関数を実行
init_printing()

# 入力の用意
input = ( Qubit("000")+Qubit("001")+Qubit("010")+Qubit("011")+Qubit("100")+Qubit("101")+\
          Qubit("110")+Qubit("111")) / sympy.sqrt(8)
input
```

```python
import numpy as np
input_np_array = 1 / np.sqrt(8) * np.ones(8)
# numpy の ifft と本節でのフーリエ変換の定義を合わせるため, sqrt(2^3)をかける
output_np_array = np.fft.ifft(input_np_array) * np.sqrt(8)
print(output_np_array)
```

```python
represent(Z(0),nqubits=1), represent(S(0),nqubits=1), represent(T(0),nqubits=1)
```

```python
QFT_gate = H(2)
QFT_gate = CGateS(1, S(2)) * QFT_gate
QFT_gate = CGateS(0, T(2)) * QFT_gate
```

```python
QFT_gate = H(1) * QFT_gate
QFT_gate = CGateS(0, S(1)) * QFT_gate
QFT_gate = H(0) * QFT_gate
```

```python
QFT_gate = SWAP(0, 2) * QFT_gate
QFT_gate
```

```python
simplify( qapply( QFT_gate * input) )
```

```python
input_2 = (Qubit("000") - Qubit("010") + Qubit("100") - Qubit("110")) / 2
simplify( qapply( QFT_gate * input_2) )
```
