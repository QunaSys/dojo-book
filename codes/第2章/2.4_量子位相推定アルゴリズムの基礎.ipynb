```python
from sympy import simplify, init_printing
from sympy.physics.quantum import represent, qapply
from sympy.physics.quantum.qubit import Qubit
from sympy.physics.quantum.gate import Z, S, T, H, CGateS, SWAP # SWAPを追加
# ベクトルや行列をきれいに表示するための関数を実行
init_printing()
# S×Tを表示
represent(S(1)*T(0), nqubits=2)
```

```python
cP_0 = CGateS(0, S(4)) * CGateS(0, T(3))
cP_1 = CGateS(1, S(4)) * CGateS(1, T(3))
cP_2 = CGateS(2, S(4)) * CGateS(2, T(3))
```

```python
PhaEst = H(2) * H(1) * H(0)
PhaEst = (cP_2*cP_2*cP_2*cP_2) * (cP_1*cP_1) * cP_0 * PhaEst
```

```python
PhaEst = SWAP(0, 2) * PhaEst
PhaEst = H(0) * PhaEst
```

```python
PhaEst = CGateS(0, S(1)) * PhaEst
PhaEst = CGateS(0, Z(1)) * PhaEst
PhaEst = H(1) * PhaEst
```

```python
PhaEst = CGateS(0, T(2)) * PhaEst
PhaEst = CGateS(0, S(2)) * PhaEst
PhaEst = CGateS(0, Z(2)) * PhaEst
PhaEst = CGateS(1, S(2)) * PhaEst
PhaEst = CGateS(1, Z(2)) * PhaEst
PhaEst = H(2) * PhaEst
```

```python
simplify( qapply( PhaEst * Qubit("11000") ) )
```

```python
simplify( qapply( PhaEst * Qubit("00000") ) ), \
simplify( qapply( PhaEst * Qubit("01000") ) ), \
simplify( qapply( PhaEst * Qubit("10000") ) ),
```

```python
input_state = 0.6 * Qubit("00000") + 0.8 * Qubit("11000")
simplify( qapply( PhaEst * input_state) )
```
