```python
import sympy
from sympy import init_printing
from sympy.physics.quantum import represent,qapply,TensorProduct
from sympy.physics.quantum.qubit import Qubit
from sympy.physics.quantum.gate import X,Y,Z,H,S,T
print(sympy.__version__)
# ベクトルや行列をきれいに表示するための関数を実行
sympy.init_printing()
```

```python
X(0)
```

```python
represent(X(0),nqubits=1) # パウリゲート X
```

```python
ket0 = Qubit('0')
S(0)*Y(0)*X(0)*H(0)*ket0
```

```python
qapply(S(0)*Y(0)*X(0)*H(0)*ket0)
```

```python
represent(qapply(S(0)*Y(0)*X(0)*H(0)*ket0))
```
