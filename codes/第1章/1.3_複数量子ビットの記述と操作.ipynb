```python
# 修正版
import sympy
from sympy import symbols
from sympy.physics.quantum import represent,qapply,TensorProduct
from sympy.physics.quantum.qubit import Qubit
from sympy.physics.quantum.gate import X,Y,Z,H,S,T,CNOT
print(sympy.__version__)
sympy.init_printing() # ベクトルや行列をきれいに表示するための関数
```

```python
a,b,c,d,e,f = symbols('alpha,beta,gamma,delta,epsilon,zeta')
psi = a*Qubit('0')+b*Qubit('1')
phi = c*Qubit('0')+d*Qubit('1')
chi = e*Qubit('0')+f*Qubit('1')

TensorProduct(psi, phi) # テンソル積
```

```python
represent(TensorProduct(psi, phi)) # 4次元ベクトルで表示
```

```python
represent(TensorProduct(chi,TensorProduct(psi, phi)))
```
