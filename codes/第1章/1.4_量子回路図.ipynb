```python
from sympy.physics.quantum import represent
from sympy.physics.quantum.gate import H,S,T,CNOT
from sympy.physics.quantum import Dagger

Toffoli = S(1) * T(0) * CNOT(0,1) * Dagger(T(1)) * \
    H(2) * CNOT(0,1) * Dagger(T(1)) * T(2) * CNOT(0,2) * Dagger(T(2)) * \
    CNOT(1,2) * T(2) * CNOT(0,2) * Dagger(T(2)) * CNOT(1,2) * H(2)
represent(Toffoli, nqubits=3)
```
