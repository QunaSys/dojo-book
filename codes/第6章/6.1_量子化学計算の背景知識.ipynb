```python
from pyscf import gto, scf

distance = 1.0 # 原子間距離（単位はオングストローム）
# 分子の情報
mole = gto.M(
    atom=[["H", [0, 0, 0]],["H", [0, 0, distance]]],
    basis="sto-3g",
    spin=0, charge=0,
)

# ハートリー・フォック計算
mf = scf.RHF(mole).run(verbose=0)
print(mf.e_tot) # ハートリー・フォックエネルギー
```

```python
from quri_parts.pyscf.mol import get_spin_mo_integrals_from_mole
from quri_parts.openfermion.mol import get_qubit_mapped_hamiltonian

# 電子積分を取り出す
full_space, mo_eints = get_spin_mo_integrals_from_mole(mole, mf.mo_coeff)
# fermion ハミルトニアンを作り, qubit ハミルトニアンに変換
ham_qubit, _ = get_qubit_mapped_hamiltonian(full_space, mo_eints)
print(ham_qubit)
```

```python
from quri_parts.openfermion.mol import get_fermionic_hamiltonian
ham_fermion = get_fermionic_hamiltonian(mo_eints)
# 表示を簡略化するためにOpenfermionのFermionOperatorに変換
import openfermion
print(openfermion.transforms.get_fermion_operator(ham_fermion))
```
