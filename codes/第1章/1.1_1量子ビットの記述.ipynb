```python
import sympy
from sympy import sqrt as sqrt
from sympy import symbols
from sympy.physics.quantum import represent
from sympy.physics.quantum.qubit import Qubit
print(sympy.__version__)
```

```python
psi = Qubit('0')
psi # ブラケット表示
```

```python
represent(psi) # 列ベクトル表示
```

```python
a, b = symbols('alpha, beta') # a,b をシンボルとして, alpha,beta として表示
ket0 = Qubit('0')
ket1 = Qubit('1')
psi = a * ket0 + b * ket1
psi # 状態をそのまま書くとケットで表示してくれる
```

```python
represent(psi)
```

```python
psi.subs([([a,1/sqrt(2)]),([b,1/sqrt(2)])]) # alpha,beta に具体的な数字を代入
```

```python
from sympy.physics.quantum.qubit import measure_all_oneshot

psi_substituted = psi.subs([([a,1/sqrt(2)]),([b,1/sqrt(2)])])
for _ in range(20):
    print(measure_all_oneshot(psi_substituted), end="")
```
