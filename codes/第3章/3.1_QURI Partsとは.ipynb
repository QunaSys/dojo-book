```python
from quri_parts.circuit import QuantumCircuit

# 3量子ビットをもつ量子回路を作成する
circuit = QuantumCircuit(3)

# 作成した量子ゲートを追加する
circuit.add_gate(X(0))

# もしくは QuantumCircuit のメソッドを用いてゲートを追加する
circuit.add_X_gate(0)
circuit.add_RX_gate(1, math.pi/3)
circuit.add_CNOT_gate(2, 1)
circuit.add_PauliRotation_gate(
    target_qubits=(0, 1, 2), pauli_id_list=(1, 2, 3), angle=math.pi/3
)
```

```python
print("Qubitcount:", circuit.qubit_count) # 量子ビット数
print("Circuitdepth:", circuit.depth) # 量子回路の深さ

# 回路中のゲートをリストとして取得する
gates = circuit.gates
print("# of gates in the circuit:", len(gates))
for gate in gates:
    print(gate)
```

```python
circuit2 = QuantumCircuit(3)
circuit2.add_Y_gate(1)
circuit2.add_H_gate(2)

# 2 つの量子回路を結合
combined = circuit + circuit2 # equivalent: combined = circuit.combine(circuit2)
print("Combined circuit:\n", combined.gates, "\n")

# 既存の量子回路を別の量子回路で拡張
circuit2 += circuit # equivalent: circuit2.extend(circuit)
print("Extended circuit:\n", circuit2.gates, "\n")

# 小さい回路をより大きな回路の中に埋め込むこともできる
circuit_larger = QuantumCircuit(5)
circuit_larger.add_X_gate(3)
circuit_smaller = QuantumCircuit(3)
circuit_smaller.add_H_gate(0)
circuit_larger.extend(circuit_smaller)
print("Circuit extended by smaller one:\n", circuit_larger.gates)
```

```python
from quri_parts.circuit.transpile import CliffordRZSetTranspiler

c_in = QuantumCircuit(3)
c_in.add_H_gate(0)
c_in.add_CNOT_gate(0, 1)
c_in.add_CNOT_gate(0, 2)
c_in.add_X_gate(0)
c_in.add_RX_gate(1, 0.25 * math.pi)
c_in.add_CNOT_gate(1, 2)
c_in.add_H_gate(0)

# 回路を変換する前の上記のコーディングそのままの回路
print(f"{c_in.depth=}, {len(c_in.gates)=}")
draw_circuit(c_in)

# CliffordRZSetTranspiler により回路形式を変形しなおした回路
transpiler = CliffordRZSetTranspiler()
c_out = transpiler(c_in)

print(f"{c_out.depth=}, {len(c_out.gates)=}")
draw_circuit(c_out)
```

```python
from quri_parts.core.operator import Operator

op = Operator(
    {
        PAULI_IDENTITY: 8 + 1j,
        pauli_label("X0 Y2"): -3
    }
)
print(op)
```

```python
# Operator に新しい項を足す
pl, coeff = pauli_label("Y0 Z3"), 10+1j
print(f"Add {coeff} * {pl}:")
op.add_term(pl, coeff)
print(op, "\n")

# Operator にすでに存在するPauliLabelを足すことで, その Operator の係数を更新する
pl, coeff = PAULI_IDENTITY, -6
print(f"Add {coeff} * {pl}:")
op.add_term(pl, coeff)
print(op, "\n")

# Operator にすでに存在するPauliLabelを足すことで, その項をキャンセルする
pl, coeff = pauli_label("X0 Y2"), 3
print(f"Add {coeff} * {pl}:")
op.add_term(pl, coeff)
print(op)
```

```python
op1 =  Operator({pauli_label("X0 Z1"): 8j})
op2 =  Operator({pauli_label("Y1"): -4})
print("op1 = ", op1)
print("op2 = ", op2)

# 足し算
print("")
print("Addition:")
print("op1 + op2", "=", op1 + op2)

# 引き算
print("")
print("Subtraction:")
print("op1 - op2", "=", op1 - op2)

# スカラー倍
print("")
print("Scalar Multiplication:")
print("op1 * 3j", "=", op1 * 3j)

# スカラーで割る計算
print("")
print("Scalar Division:")
print("op1 / 2j", "=", op1 / 2j)

# 演算子どうしの掛け算
print("")
print("Operator Multiplication:")
print("op1 * op2", "=", op1 * op2)
print("op2 * op1", "=", op2 * op1)

# エルミート共役
print("")
print("Hermitian Conjugation:")
print("op1^†", "=", op1.hermitian_conjugated())
print("op2^†", "=", op2.hermitian_conjugated())
```

```python
import numpy as np
from quri_parts.circuit import LinearMappedUnboundParametricQuantumCircuit, CONST

linear_param_circuit = LinearMappedUnboundParametricQuantumCircuit(2)
theta, phi = linear_param_circuit.add_parameters("theta", "phi")

linear_param_circuit.add_H_gate(0)
linear_param_circuit.add_CNOT_gate(0, 1)
linear_param_circuit.add_ParametricRX_gate(0, {theta: 1/2, phi: 1/3, CONST: np.pi/2})
linear_param_circuit.add_ParametricRY_gate(0, {theta: -1/2, phi: 1/3})
linear_param_circuit.add_ParametricRZ_gate(1, {theta: 1/3, phi: -1/2, CONST: -np.pi/2})
```

```python
from quri_parts.core.state import quantum_state

# ParametricCircuitQuantumState
param_circuit_state = quantum_state(n_qubits=2, circuit=linear_param_circuit)
print(param_circuit_state)

# ParametricQuantumStateVector
param_state_vector = quantum_state(
    n_qubits=2, circuit=linear_param_circuit, vector=np.array([0.5, -0.5, 0.5j, -0.5j])
)
print(param_state_vector)
```

```python
params = [0.1, 0.2]

# ParametricCircuitQuantumState -> GeneralCircuitQuantumState
circuit_state = param_circuit_state.bind_parameters(params)

# ParametricQuantumStateVector -> QuantumStateVector
vector_state = param_state_vector.bind_parameters(params)
```

```python
from quri_parts.core.state import quantum_state
from quri_parts.qulacs.estimator import create_qulacs_vector_estimator

# パラメータを代入した量子回路を使って量子状態を生成
circuit = linear_param_circuit.bind_parameters([0.2, 0.3])
circuit_state = quantum_state(2, circuit=circuit)

from quri_parts.core.operator import Operator, pauli_label
op = Operator({
    pauli_label("X0 Y1"): 0.5 + 0.5j,
    pauli_label("Z0 X1"): 0.2,
})

estimator = create_qulacs_vector_estimator()

estimate = estimator(op, circuit_state)
print(f"Estimated expectation value: {estimate.value}")
```

```python
from quri_parts.qulacs.estimator import (
    create_qulacs_vector_parametric_estimator,
    create_qulacs_vector_concurrent_parametric_estimator,
)

parametric_estimator = create_qulacs_vector_parametric_estimator()
concurrent_parametric_estimator = create_qulacs_vector_concurrent_parametric_estimator()
```

```python
# parametric_circuit_stateの期待値推定
print("Estimate with a set of circuit parameter:")
print(parametric_estimator(op, param_circuit_state, [0.1, 0.2]))

# parametric_circuit_stateの複数パラメータセットにおける期待値推定
print("")
print("Concurrent Estimate with two sets of circuit parameters:")
print(concurrent_parametric_estimator(op, param_circuit_state, [[0.1, 0.2], [0.3, 0.4]]))
```
